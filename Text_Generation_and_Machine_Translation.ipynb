2. Explain the role of probabilistic modeling in generative models. How do
these models differ from discriminative models?

ANS- Role of Probabilistic Modeling in Generative Models:
Probabilistic modeling is fundamental to generative AI. At its core, a generative model learns the underlying probability distribution of the training data. This means it tries to understand how the data was generated.

Imagine you have a dataset of cat images. A generative model doesn't just learn to recognize a cat; it learns the statistical properties that make an image a "cat image." It learns the distribution of pixel values, shapes, textures, and patterns that constitute a cat.

The key roles are:

Learning Data Distribution: The primary goal is to model the joint probability distribution P(X, Y) (or P(X) if unsupervised), where X represents the input data (e.g., image pixels) and Y represents the labels (e.g., "cat"). By learning this distribution, the model can understand the inherent structure and relationships within the data.
Generating New Samples: Once P(X) (or P(X, Y)) is learned, the model can sample from this distribution to create entirely new, unseen data that resembles the training data. This is why they are called generative models – they generate new content.
Handling Missing Data: By understanding the full data distribution, generative models can be used to impute missing values or fill in gaps in data.
Anomaly Detection: Data points that have a very low probability under the learned distribution P(X) can be identified as anomalies or outliers.
Examples of generative models include Variational Autoencoders (VAEs), Generative Adversarial Networks (GANs), and Large Language Models (LLMs) like GPT, which learn the probability distribution of sequences of words.

Generative Models vs. Discriminative Models:
The distinction lies in what they aim to model:

Generative Models:

Goal: Learn the joint probability distribution P(X, Y) or the data distribution P(X). Essentially, they model how the data is generated.
Output: Can generate new data samples that are similar to the training data. They can also be used for classification by calculating P(Y|X) from P(X, Y).
Examples: GANs, VAEs, Naive Bayes, Hidden Markov Models, GPT (for text generation).
Strengths: Better for generating new data, detecting anomalies, and handling missing data. Can perform classification.
Weaknesses: Often more complex to train and require more data.
Discriminative Models:

Goal: Learn the conditional probability distribution P(Y|X). They directly model the decision boundary between classes, aiming to discriminate between different categories.
Output: Predicts the label Y given input X. They cannot generate new data.
Examples: Logistic Regression, Support Vector Machines (SVMs), Perceptrons, most standard Neural Networks (for classification tasks).
Strengths: Typically simpler to train, require less data, and often achieve higher accuracy for classification tasks.
Weaknesses: Cannot generate new data, less robust to changes in input distribution, and cannot easily handle missing data.
Analogy:

Imagine you want to distinguish between images of cats and dogs:

A discriminative model learns what distinguishes a cat from a dog (e.g., pointy ears vs. floppy ears) and draws a boundary. It's like a bouncer at a club deciding if someone is a cat or a dog based on a few key features.
A generative model learns to draw a cat and learns to draw a dog. If it can generate a convincing cat or dog, it implicitly understands what makes them different. It's like an artist who can draw both cats and dogs, and therefore understands the essence of each.


3. : What is the difference between Autoencoders and Variational Autoencoders (VAEs) in the context of text generation?

ANS-Autoencoders (AEs) for Text Generation:
Concept: An Autoencoder is a type of neural network that learns an efficient data encoding (a compressed representation) in an unsupervised manner. It consists of two main parts:

Encoder: Maps the input data to a lower-dimensional latent space representation (the encoding or bottleneck).
Decoder: Reconstructs the input data from this latent space representation.
In Text Generation: For text, an AE would typically take an input sequence (e.g., a sentence), encode it into a fixed-size vector (the latent representation), and then try to decode that vector back into the original sentence. The goal is to learn a meaningful representation of the text.

Limitations for Text Generation:

No Generative Guarantees: The latent space learned by a standard AE is not necessarily continuous or easily navigable. If you sample a random point from this latent space and pass it through the decoder, there's no guarantee that the output will be a coherent or meaningful text. The latent space is often discontinuous, meaning that small interpolations between two valid encodings might lead to invalid or nonsensical outputs.
Reconstruction, Not Generation: AEs are primarily designed for dimensionality reduction and data reconstruction, not for generating novel, coherent samples.
Mode Collapse Risk: If used for generation, they might struggle to produce diverse outputs, often collapsing to produce only a few typical examples.
Variational Autoencoders (VAEs) for Text Generation:
Concept: VAEs extend traditional AEs by introducing a probabilistic approach to the encoder and decoder. Instead of mapping an input to a single point in the latent space, the encoder of a VAE maps it to parameters of a probability distribution (typically a Gaussian distribution) over the latent space. The decoder then samples from this learned distribution to reconstruct the input.

Key Differences/Improvements:

Probabilistic Encoder/Decoder:
AE: Encoder outputs a single latent vector z. Decoder takes z to reconstruct X.
VAE: Encoder outputs mean μ and variance σ (or log-variance) for a distribution q(z|X). The latent vector z is then sampled from this distribution. The decoder then reconstructs X from the sampled z.
Regularized Latent Space: VAEs introduce a regularization term (KL-divergence) in their loss function. This term forces the learned latent distributions q(z|X) to be close to a prior distribution (usually a standard normal distribution, p(z)). This regularization ensures that:
The latent space is continuous and smooth. This is crucial because it means that interpolating between two valid latent vectors will likely result in another valid and meaningful point, which can be decoded into a coherent output.
The latent space is structured. Similar inputs will map to similar regions in the latent space, and different regions correspond to different features or characteristics of the data.
The model can generate new samples by simply sampling from the prior distribution p(z) (e.g., a standard normal distribution) and passing these samples through the decoder.
In Text Generation: For text, a VAE would learn a continuous and structured latent representation of sentences. To generate new text:

You can sample a vector z from a simple prior distribution (e.g., a standard normal distribution).
Pass this z through the decoder.
The decoder, often a Recurrent Neural Network (RNN) or Transformer, will then generate a new, coherent sentence that reflects the characteristics encoded in z.
Advantages for Text Generation:

Generative Capability: VAEs are explicitly designed for generation due to their continuous and regularized latent space. You can sample arbitrary points from the latent space and expect meaningful outputs.
Meaningful Interpolation: You can interpolate between two latent representations of different sentences to generate text that smoothly transitions between the two.
Controlled Generation: By manipulating specific dimensions in the latent space, you can potentially control attributes of the generated text (e.g., sentiment, topic, style).




4. Describe the working of attention mechanisms in Neural Machine Translation (NMT). Why are they critical?

ANS-Working of Attention Mechanisms in NMT:
Traditionally, NMT models relied on a single encoder-decoder architecture where the encoder compressed the entire source sentence into a fixed-size 'context vector'. This vector was then passed to the decoder to generate the target sentence. The problem with this approach was that for long sentences, this single context vector became a bottleneck, struggling to retain all the necessary information.

Attention mechanisms address this bottleneck by allowing the decoder to 'look back' at the entire source sentence (or its encoded representations) during each step of decoding. Instead of relying on a single, fixed-size context vector, attention allows the model to dynamically weigh the importance of different parts of the source sentence when generating each word of the target sentence.

Here's a step-by-step breakdown:

Encoder: The encoder (often a Recurrent Neural Network like an LSTM or GRU, or a Transformer's encoder stack) processes the input (source) sentence, word by word, and generates a sequence of hidden states (or annotations/representations) for each word. Each hidden state h_i contains information about the i-th word and its context in the source sentence.

Decoder Initialization: The decoder (another RNN or Transformer's decoder stack) starts generating the target sentence. For the first output word, or after generating a previous word, the decoder has its own hidden state s_t.

Alignment Score Calculation: For each decoding step t, the decoder's current hidden state s_t is compared with each of the encoder's hidden states h_i (for i from 1 to source_length). This comparison produces an "alignment score" (or "energy score" or "attention weight") e_ti. A higher score indicates that the i-th source word is more relevant for generating the current target word. Common ways to calculate this include dot products, additive (Bahdanau) attention, or multiplicative (Luong) attention.

Softmax Normalization: The alignment scores e_ti are then normalized using a softmax function to get a set of "attention weights" α_ti. These weights sum to 1 and represent a probability distribution over the source words. α_ti tells us how much attention the decoder should pay to the i-th source word when generating the t-th target word.

Context Vector Creation: A context vector c_t is computed as a weighted sum of the encoder's hidden states, where the weights are the attention weights α_ti. So, c_t = Σ (α_ti * h_i). This context vector c_t captures the relevant information from the source sentence for generating the current target word.

Target Word Prediction: The context vector c_t, along with the decoder's current hidden state s_t (and sometimes the previously predicted target word), is used by the decoder to predict the next word in the target sequence.

This process repeats for each word generated in the target sentence, allowing the model to focus on different parts of the source sentence as needed.

Why are Attention Mechanisms Critical?
Attention mechanisms are critical for several reasons:

Resolving the Fixed-Size Context Vector Bottleneck: As mentioned, traditional encoder-decoder models struggled with long sentences because the single context vector couldn't store all the necessary information. Attention allows the model to access all encoder hidden states directly, bypassing this information loss.

Handling Long-Range Dependencies: In translation, a word in the target sentence might depend on a word far away in the source sentence. Attention provides a direct connection between these distant words, making it easier to capture these long-range dependencies than traditional RNNs which suffer from vanishing/exploding gradients over long sequences.

Improved Translation Quality: By enabling the model to selectively focus on relevant parts of the source sentence, attention leads to significantly more accurate and fluent translations, especially for complex sentences and longer texts.

Interpretability/Alignment: The attention weights α_ti provide a degree of interpretability. By visualizing these weights, we can see which source words the model is 'attending' to when generating each target word. This often reveals a meaningful alignment between source and target words, which is intuitive for humans and provides insight into the model's decision-making process.

Foundation for the Transformer Architecture: Attention, particularly the concept of "self-attention" (where attention is applied within the same sequence), is the core innovation behind the Transformer architecture. Transformers, which rely heavily on multi-head self-attention, have become the dominant architecture in NMT and many other NLP tasks, proving the critical importance and versatility of attention.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

### 1. Preprocess the data (tokenize and pad the sequences)

```python
# Define the dataset
data = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]

# Initialize tokenizer
# num_words=None means it will consider all words
# oov_token='<unk>' handles out-of-vocabulary words
tokenizer = Tokenizer(num_words=None, oov_token='<unk>')
tokenizer.fit_on_texts(data)

# Convert text to sequences of integers
sequences = tokenizer.texts_to_sequences(data)

# Get word-to-index and index-to-word mappings
word_to_index = tokenizer.word_index
index_to_word = {v: k for k, v in word_to_index.items()}

# Add padding token to index_to_word if not already present
if 0 not in index_to_word:
    index_to_word[0] = '<pad>'

# Pad sequences to a uniform length
# We'll use the maximum length found in the dataset for simplicity
max_sequence_length = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_sequence_length, padding='post')

vocab_size = len(tokenizer.word_index) + 1  # +1 for padding/0-index

print(f"Original Data: {data}")
print(f"\nMax Sequence Length: {max_sequence_length}")
print(f"Vocabulary Size: {vocab_size}")
print(f"\nExample Sequence (first sentence): {sequences[0]}")
print(f"Padded Example Sequence (first sentence): {padded_sequences[0]}")
print(f"\nWord to Index: {dict(list(word_to_index.items())[:10])}...") # Displaying first 10 for brevity
print(f"Index to Word: {dict(list(index_to_word.items())[:10])}...") # Displaying first 10 for brevity
```

### 2. Build a basic VAE model for text reconstruction

```python
class Sampling(layers.Layer):
    """Uses (z_mean, z_log_var) to sample z, the vector encoding a digit."""
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

class VAE(keras.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super(VAE, self).__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.total_loss_tracker = keras.metrics.Mean(name="total_loss")
        self.reconstruction_loss_tracker = keras.metrics.Mean(name="reconstruction_loss")
        self.kl_loss_tracker = keras.metrics.Mean(name="kl_loss")

    @property
    def metrics(self):
        return [
            self.total_loss_tracker,
            self.reconstruction_loss_tracker,
            self.kl_loss_tracker,
        ]

    def train_step(self, data):
        if isinstance(data, tuple):
            data = data[0] # Assuming data is (x, y) where x=y for reconstruction

        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)

            # Reconstruction loss: sparse categorical crossentropy for sequence prediction
            # We use `sparse_categorical_crossentropy` because our targets are integers (word indices).
            # `from_logits=True` if the decoder outputs raw logits, `False` if it outputs probabilities (like softmax).
            # Here we assume the decoder outputs logits.
            reconstruction_loss = tf.reduce_mean(
                keras.losses.sparse_categorical_crossentropy(data, reconstruction, from_logits=True)
            )

            # KL divergence loss
            kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
            kl_loss = tf.reduce_mean(tf.reduce_sum(kl_loss, axis=1))

            # Total loss is reconstruction loss + KL divergence loss (weighted)
            # The weight for KL loss can be tuned. For simplicity, we'll use 1.
            total_loss = reconstruction_loss + kl_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        self.total_loss_tracker.update_state(total_loss)
        self.reconstruction_loss_tracker.update_state(reconstruction_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        return {
            "loss": self.total_loss_tracker.result(),
            "reconstruction_loss": self.reconstruction_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }

# Model parameters
embedding_dim = 64
latent_dim = 32
hidden_dim = 128

# Encoder
encoder_inputs = keras.Input(shape=(max_sequence_length,))
x = layers.Embedding(vocab_size, embedding_dim, name="embedding_layer")(encoder_inputs)
x = layers.GRU(hidden_dim, return_sequences=False)(x) # return_sequences=False for a single vector output
z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
z = Sampling()([z_mean, z_log_var])
encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")
encoder.summary()

# Decoder
latent_inputs = keras.Input(shape=(latent_dim,))
x = layers.RepeatVector(max_sequence_length)(latent_inputs) # Repeat latent vector for each time step
x = layers.GRU(hidden_dim, return_sequences=True)(x)
decoder_outputs = layers.TimeDistributed(layers.Dense(vocab_size))(x) # Output a probability distribution over vocab for each step
decoder = keras.Model(latent_inputs, decoder_outputs, name="decoder")
decoder.summary()

# VAE model
vae = VAE(encoder, decoder)
vae.compile(optimizer=keras.optimizers.Adam())
```

### 3. Train the model and show how it reconstructs or generates similar sentences

```python
# Train the VAE
print("\nTraining VAE...")
vae.fit(padded_sequences, epochs=50, batch_size=len(data)) # Batch size as length of data for small dataset

# Function to decode sequences back to text
def sequences_to_text(sequences_array):
    texts = []
    for seq in sequences_array:
        # Convert indices to words, ignoring padding (0) and unknown token
        words = [index_to_word[idx] for idx in seq if idx != 0 and idx != word_to_index.get('<unk>', -1)]
        texts.append(' '.join(words))
    return texts

# --- Reconstruction ---
print("\n--- Reconstruction Examples ---")
# Encode the input sequences
z_mean, z_log_var, z = vae.encoder.predict(padded_sequences)

# Decode the latent vectors to reconstruct the original sentences
reconstructed_logits = vae.decoder.predict(z)
reconstructed_sequences = np.argmax(reconstructed_logits, axis=-1)

original_texts = sequences_to_text(padded_sequences)
reconstructed_texts = sequences_to_text(reconstructed_sequences)

for i in range(len(data)):
    print(f"Original:      {original_texts[i]}")
    print(f"Reconstructed: {reconstructed_texts[i]}\n")

# --- Generation ---
print("\n--- Generation Examples ---")
num_generation_samples = 3
# Generate new sentences by sampling from the latent space (prior distribution, e.g., standard normal)
samples = tf.random.normal(shape=(num_generation_samples, latent_dim))
generated_logits = vae.decoder.predict(samples)
generated_sequences = np.argmax(generated_logits, axis=-1)

generated_texts = sequences_to_text(generated_sequences)

for i, text in enumerate(generated_texts):
    print(f"Generated Sentence {i+1}: {text}")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

print("TensorFlow Version:", tf.__version__)
```

### 1. & 2. Data Preparation and Preprocessing

```python
# Small English-Spanish dataset
text_pairs = [
    ("go .", "vete ."),
    ("go .", "vayase ."),
    ("go away .", "vete ."),
    ("go away .", "vaya ."),
    ("wash your hands .", "lavate las manos ."),
    ("come here .", "ven aqui ."),
    ("i love you .", "te amo ."),
    ("i love you .", "te quiero ."),
    ("hello .", "hola ."),
    ("how are you ?", "como estas ?"),
    ("i am fine .", "estoy bien ."),
    ("thank you .", "gracias ."),
    ("please .", "por favor ."),
    ("good morning .", "buenos dias ."),
    ("good night .", "buenas noches .")
]

# Add start and end tokens to target sentences
source_texts = [pair[0] for pair in text_pairs]
target_texts = ["startseq " + pair[1] + " endseq" for pair in text_pairs]

# Tokenize English (source) sentences
eng_tokenizer = Tokenizer(filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~\t\n', lower=True)
eng_tokenizer.fit_on_texts(source_texts)
eng_sequences = eng_tokenizer.texts_to_sequences(source_texts)
eng_word_index = eng_tokenizer.word_index

# Tokenize Spanish (target) sentences
spa_tokenizer = Tokenizer(filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~\t\n', lower=True)
spa_tokenizer.fit_on_texts(target_texts)
spa_sequences = spa_tokenizer.texts_to_sequences(target_texts)
spa_word_index = spa_tokenizer.word_index

# Get index-to-word mappings for decoding
eng_index_to_word = {v: k for k, v in eng_word_index.items()}
spa_index_to_word = {v: k for k, v in spa_word_index.items()}

# Padding
max_eng_len = max(len(seq) for seq in eng_sequences)
max_spa_len = max(len(seq) for seq in spa_sequences)

encoder_input_data = pad_sequences(eng_sequences, maxlen=max_eng_len, padding='post')
decoder_input_data = pad_sequences(spa_sequences, maxlen=max_spa_len, padding='post')

# Decoder target data needs to be one-hot encoded or sparse categorical crossentropy
# We'll create target sequences for teacher forcing: shifted one position to the left
decoder_target_data = np.zeros_like(decoder_input_data)
decoder_target_data[:, :-1] = decoder_input_data[:, 1:]

eng_vocab_size = len(eng_word_index) + 1
spa_vocab_size = len(spa_word_index) + 1

print(f"English Vocab Size: {eng_vocab_size}")
print(f"Spanish Vocab Size: {spa_vocab_size}")
print(f"Max English Sentence Length: {max_eng_len}")
print(f"Max Spanish Sentence Length: {max_spa_len}")
print(f"\nExample Encoder Input: {encoder_input_data[0]}")
print(f"Example Decoder Input: {decoder_input_data[0]}")
print(f"Example Decoder Target: {decoder_target_data[0]}")
```

### 3., 4., 5. & 6. Build Encoder, Attention, Decoder, and Model

```python
# Model Parameters
embedding_dim = 256
hidden_units = 512

# Encoder
encoder_inputs = keras.Input(shape=(max_eng_len,))
encoder_embedding = layers.Embedding(eng_vocab_size, embedding_dim)(encoder_inputs)
encoder_outputs, encoder_state = layers.GRU(hidden_units, return_state=True)(encoder_embedding)

encoder_model = keras.Model(encoder_inputs, [encoder_outputs, encoder_state], name="encoder")
encoder_model.summary()


# Decoder
decoder_inputs = keras.Input(shape=(max_spa_len,))
decoder_embedding = layers.Embedding(spa_vocab_size, embedding_dim)(decoder_inputs)

# Attention Layer
class BahdanauAttention(layers.Layer):
    def __init__(self, units):
        super(BahdanauAttention, self).__init__()
        self.W1 = layers.Dense(units)
        self.W2 = layers.Dense(units)
        self.V = layers.Dense(1)

    def call(self, query, values):
        # query hidden state shape == (batch_size, hidden_size)
        # values shape == (batch_size, max_len, hidden_size)

        # we are doing this to broadcast addition along the time axis to calculate the score
        query_with_time_axis = tf.expand_dims(query, 1) # (batch_size, 1, hidden_size)

        # score shape == (batch_size, max_len, 1)
        # we get 1 at the last axis because we are applying score to self.V
        # the shape of the tensor before applying self.V is (batch_size, max_len, units)
        score = self.V(tf.nn.tanh(self.W1(query_with_time_axis) + self.W2(values)))

        # attention_weights shape == (batch_size, max_len, 1)
        attention_weights = tf.nn.softmax(score, axis=1)

        # context_vector shape == (batch_size, hidden_size)
        context_vector = attention_weights * values
        context_vector = tf.reduce_sum(context_vector, axis=1)

        return context_vector, attention_weights

attention_layer = BahdanauAttention(hidden_units)

# Define the decoder GRU layer
decoder_gru = layers.GRU(hidden_units, return_sequences=True, return_state=True)
decoder_dense = layers.Dense(spa_vocab_size)

# Create the training decoder model
# Takes encoder_outputs (all hidden states) and initial_state (last hidden state of encoder)
def create_decoder_model(encoder_outputs, initial_state):
    current_state = initial_state
    all_outputs = []

    # Iterate through the decoder input sequence (for teacher forcing)
    for t in range(max_spa_len):
        decoder_single_step_input = decoder_embedding[:, t:t+1, :]

        # Calculate attention for the current decoder state and all encoder outputs
        context_vector, _ = attention_layer(current_state, encoder_outputs)

        # Concatenate context vector with decoder input
        decoder_input_with_attention = tf.concat([tf.expand_dims(context_vector, 1), decoder_single_step_input], axis=-1)

        # Pass through GRU
        output, current_state = decoder_gru(decoder_input_with_attention, initial_state=current_state)
        output = decoder_dense(output)
        all_outputs.append(output)

    decoder_outputs = tf.concat(all_outputs, axis=1)
    return decoder_outputs, current_state

# Full Seq2Seq model for training
# Encoder outputs all hidden states and final state
encoder_all_outputs, encoder_final_state = layers.GRU(hidden_units, return_sequences=True, return_state=True, name="encoder_gru")(encoder_embedding)

# Decoder for training
decoder_outputs_train, _ = create_decoder_model(encoder_all_outputs, encoder_final_state)

model = keras.Model([encoder_inputs, decoder_inputs], decoder_outputs_train)
model.compile(optimizer=keras.optimizers.Adam(), loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True))
model.summary()
```

### 6. Train the Model

```python
history = model.fit(
    [encoder_input_data, decoder_input_data],
    decoder_target_data,
    batch_size=32,
    epochs=100,
    validation_split=0.2
)
```

### 7. Translate (Inference Model)

```python
# Inference Encoder (same as training encoder)
inference_encoder_model = keras.Model(encoder_inputs, [encoder_all_outputs, encoder_final_state])

# Inference Decoder
decoder_state_input = keras.Input(shape=(hidden_units,))

# Context vector for attention
encoder_output_inf = keras.Input(shape=(max_eng_len, hidden_units))
context_vector_inf, attention_weights_inf = attention_layer(decoder_state_input, encoder_output_inf)

decoder_input_single_step = keras.Input(shape=(1,))
decoder_embedding_inf = layers.Embedding(spa_vocab_size, embedding_dim)(decoder_input_single_step)
decoder_input_with_attention_inf = tf.concat([tf.expand_dims(context_vector_inf, 1), decoder_embedding_inf], axis=-1)

decoder_output_inf, decoder_state_inf = decoder_gru(decoder_input_with_attention_inf, initial_state=decoder_state_input)
decoder_output_inf = decoder_dense(decoder_output_inf)

inference_decoder_model = keras.Model(
    [decoder_input_single_step, decoder_state_input, encoder_output_inf],
    [decoder_output_inf, decoder_state_inf, attention_weights_inf]
)


def translate_sentence(input_sentence):
    input_sequence = eng_tokenizer.texts_to_sequences([input_sentence])
    padded_input_sequence = pad_sequences(input_sequence, maxlen=max_eng_len, padding='post')

    # Encode the input sentence
    enc_outputs, states_value = inference_encoder_model.predict(padded_input_sequence)

    # Initialize decoder input with 'startseq'
    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = spa_word_index['startseq']

    stop_condition = False
    decoded_sentence = ''

    while not stop_condition:
        output_tokens, states_value, attention_weights = inference_decoder_model.predict(
            [target_seq, states_value, enc_outputs])

        # Sample a token
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_word = spa_index_to_word.get(sampled_token_index, '') # Use .get with default empty string

        if sampled_word == 'endseq' or len(decoded_sentence.split()) >= max_spa_len - 1:
            stop_condition = True
        else:
            if sampled_word != '': # Only append if it's a valid word
                decoded_sentence += ' ' + sampled_word

        # Update the target sequence (of length 1).
        target_seq = np.zeros((1, 1))
        target_seq[0, 0] = sampled_token_index

    return decoded_sentence.strip()

# --- Demonstration ---
print("\n--- Translation Demonstration ---")

# Test cases from our small dataset
example_sentences = [
    "go .",
    "wash your hands .",
    "how are you ?",
    "thank you .",
    "good morning ."
]

for eng_sentence in example_sentences:
    spa_translation = translate_sentence(eng_sentence)
    print(f"English: {eng_sentence:<20} -> Spanish: {spa_translation}")

# Test with a new sentence (might not be perfect due to small dataset)
new_eng_sentence = "hello how are you ?"
new_spa_translation = translate_sentence(new_eng_sentence)
print(f"\nEnglish: {new_eng_sentence:<20} -> Spanish: {new_spa_translation}")
```

```python
from transformers import pipeline, set_seed

# Set a seed for reproducibility
set_seed(42)

# Initialize the text generation pipeline using GPT-2
generator = pipeline('text-generation', model='gpt2')

# The poetry dataset provided
poetry_dataset = [
    "Roses are red, violets are blue,",
    "Sugar is sweet, and so are you.",
    "The moon glows bright in silent skies,",
    "A bird sings where the soft wind sighs."
]

# Craft a prompt that simulates the poetic structure and language
# We'll use the first two lines as a stylistic starting point.
prompt = "Roses are red, violets are blue,\nSugar is sweet, and so are you.\nA soft breeze blows," # The prompt to guide the generation

print(f"Prompt used: '{prompt}'\n")

# Generate a poem
# num_return_sequences: number of poems to generate
# max_new_tokens: maximum length of the generated text
# temperature: controls randomness (higher = more random)
# top_k: filters top_k most likely words
# early_stopping: stop generation when all sequences are finished
generated_poems = generator(
    prompt,
    max_new_tokens=20, # Generate 20 new tokens after the prompt
    num_return_sequences=1,
    temperature=0.8,
    top_k=50,
    pad_token_id=generator.tokenizer.eos_token_id # Avoid warning message
)

# Extract and print the generated poem
# We take the first generated sequence and clean it up to get 2-4 lines
full_generated_text = generated_poems[0]['generated_text']

# Split the text into lines and take the relevant ones
lines = full_generated_text.split('\n')

# Filter out empty lines and take 2 to 4 lines, ensuring it doesn't just repeat the prompt
# We skip the first two lines as they are part of the prompt
poem_lines = [line.strip() for line in lines[2:] if line.strip()]

# If not enough new lines, we might need to adjust slicing or max_new_tokens
if len(poem_lines) < 2:
    # Fallback to include more if not enough new ones were generated
    poem_lines = [line.strip() for line in lines[len(poetry_dataset):] if line.strip()]

# Ensure we have at least 2 and at most 4 lines from the generated part
final_poem = "\n".join(poem_lines[:4]) if len(poem_lines) >=2 else "\n".join(lines[2:6]) # Fallback to a fixed range if filtering is too aggressive

print("--- Generated Poem ---")
print(final_poem)
```
